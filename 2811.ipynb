```python

import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/content/data_training.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.isna().sum()
```

```python
df.dtypes
```

```python
number = LabelEncoder()
df['Jenis_Kelamin'] = number.fit_transform(df['Jenis_Kelamin'])
df['Jenis_Seleksi'] = number.fit_transform(df['Jenis_Seleksi'])
df['Pendidikan_Ibu'] = number.fit_transform(df['Pendidikan_Ibu'])
df['IP_SMT1'] = number.fit_transform(df['IP_SMT1'])
df['IP_SMT2'] = number.fit_transform(df['IP_SMT2'])
df['IP_SMT3'] = number.fit_transform(df['IP_SMT3'])
df['IP_SMT4'] = number.fit_transform(df['IP_SMT4'])
df
```

```python
df_train = df[0:5]
```

```python
df_train
```

```python
df_test = df[5:]
```

```python
df_test
```

```python
X_train = df_train.iloc[:,1:3]
y_train = df_train.iloc[:,3:4]
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
X_train
```

```python
y_train
```

```python
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
X_test = df_test.iloc[:,1:3]
X_test
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
pred_proba = model.predict_proba(X_test)
```

```python
pred_proba
```
